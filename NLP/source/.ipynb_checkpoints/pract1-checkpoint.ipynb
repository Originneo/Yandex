```python
import os
import sys
import pandas as pd
import fnmatch
import json
import re
```

```python
data = pd.read_csv('../data/prepared.csv')
data.text = data.text.apply(lambda x: [int(i) for i in x.split()])
data.head()
```

```python
num2v = json.loads(open('../data/num2v.json').read())
v2num = json.loads(open('../data/v2num.json').read())
```

```python
dic = []
for i in num2v:
    dic.append([i, num2v[i]])
dic = pd.DataFrame(dic)
dic.columns = ['index', 'word']
dic.head()
```

```python
n1 = 10
n2 = 10
N = 20

import numpy as np


ai = []
for indx in dic.index:
    a = [0, 0]
    for i, tag in enumerate(list(set(data.tag))):
        for cur_tag, text in data.as_matrix():
            if tag == cur_tag and indx in text:
                a[i] += 1
    ai.append([a[0]/n1, a[1]/n2])
dic['a'] = np.array(ai)[:, 0]
dic['b'] = np.array(ai)[:, 1]
dic.head()
```

```python
import math
const = 0.5

dic['c'] = dic[['a', 'b']].apply(lambda x: math.log((x[0]*(1 - x[1]) + const)/(x[1]*(1 - x[0]) + const)), axis=1)
dic.sort_values('c')
```

```python
dic.to_csv('../data/stats.csv', index=False)
```

```python
C = 0

for a, b in dic[['a', 'b']].as_matrix():
    C += math.log((1 - a + const)/(1 - b + const))
C
```
